## Algoritmo Covarianza

```python
#Algoritmo que devuelve la matriz de covarianza en base a un conjunto de datos presentado en una matriz(DataFrame)
def matrizCovarianza(data):
    #el tamaño de la muestra
    tam = len(data.index)
    #la matriz de promedios
    dim = len(data.T.index)
    #la dimension de cada elemento de la muestra
    cov = np.zeros((dim, dim))
    #promedio vectorial
    mean = data.mean(axis = 0)
    #variables de iteración
    i = j = k = 0
    #itera sobre una dimension de la matriz
    for i in range(dim):
        #itera sobre la otra dimension de la matriz, hasta la diagonal
        #(aprovechando la simetría de la matriz de covarianza)
        for j in range(0,i+1):
            suma = 0
            #Calcula la sumatoria que determina el término i,j de la matriz
            for k in range(tam):
                a = (data.ix[k,i] - mean[i])*(data.ix[k,j] - mean[j])
                suma += a
            #asigna los valores a los cuadros diagonalmente simétricos de la matriz de covarianza    
            cov[j][i] = suma/(tam-1)
            cov[i][j] = suma/(tam-1)

    return cov
```

## Matriz de covarianza del problema

```python
#matriz de covarianza de los datos de tasas de homicidio de los departamentos de Colombia a través de los años
cova = matrizCovarianza(TasasHom)
```

## Cálculo de autovalores y autovectores

```python
#saca los autovalores y autovectores de la matriz de covarianza
eig = np.linalg.eig(cova)
#valores propios con un numero asociado que indica la posición en la lista su autovector correspondiente
eigenValuesN = np.array([np.array([(eig[0])[k] , k]) for k in range(len(eig[0]))])
#Se ordena el array respecto a la columna de autovalores (de mayor a menor)
eigenValuesN = -eigenValuesN
eigenValuesN = -eigenValuesN[[eigenValuesN[:,0].argsort()]]
```

## Reducción de la dimensionalidad

Se tendrá en cuenta el criterio del 90%

Esto es, solo se analizarán los primeros n autovalores cuya sumatoria sea el 90% del total

```python
#calcula el peso total de los auto valores (se escogerán los primeros n que sean el 90% del peso total)
pesoT = sum(eigenValuesN.T[0])
index = 0
ans = 0
while ans < 0.9:
    ans+= eigenValuesN[index][0]/pesoT
    print ans
    index += 1
#El index paró un índice atrás (el while tarda 1 index en reconocer que ya no hay ciclo)
index = index-1 
```

```python
#valores propios que se van a tomar
eigenValuesN = eigenValuesN[:index+1]
print "EigenValues: \n"
print eigenValuesN
#vectores propios correspondientestn
eigenVects = (eig[1])[eigenValuesN.T[1].astype(int)]
print "\n EigenVectors: \n"
print eigenVects
#espacio de indices de los autovectores
linsp = eigenValuesN.T[1].astype(int)
linsp
```

# Clustering, K-Means

```python
#Pasa el dataframe del historial a array para poder aplicar los algoritmos
arrayHistorial = np.array(TasasHom)
```

```python
#algoritmo para buscar clusters en una muestra dada
def clustering( nClusters, data ):
    i = j = 0
    dim = len(data[0])
    muestra = len(data)
    indexClusters = -np.ones(muestra)
    clusterPoints = 10*np.random.random((nClusters, dim))
    #clusterPoints = np.array([[0,0],[100,100]])
    termino = False
    while termino == False :
        hayClustersVacios = False
        indexClusters = -np.ones(muestra)
        i = 0
        for i in range(muestra):
            dmin = np.linalg.norm(data[i] - clusterPoints[0])
            nmin = 0
            for j in range(1, nClusters):
                n = np.linalg.norm(data[i] - clusterPoints[j])
                if n < dmin:
                    dmin = n
                    nmin = j
            indexClusters[i] = nmin
        
        newClusters = np.zeros((nClusters,dim))
        k = 0
        for k in range(nClusters):
            index = where(indexClusters == k)
            if len(index[0]) != 0 :
                newClusters[k] = np.mean(data[index], linspace(0,dim,1))
            else: 
                hayClustersVacios = True
        if (newClusters == clusterPoints).all():
            #c = where((newClusters == [0.,0.]).all())
            #print "\n", newClusters, c
            #if len(c[0]) != 0:
             #   newClusters = 10*np.random.random((nClusters, dim))
            #else:
            if hayClustersVacios == True:
                newClusters = 10*np.random.random((nClusters, dim))
            else:    
                termino = True
        #print indexClusters
        #print clusterPoints
        clusterPoints = newClusters
    k = 0       
    return [[data[where(indexClusters == k)] for k in range(nClusters)], clusterPoints]
        
```

```python
#Calcula la desviación estandar de un cluster respecto a su centro
def distProm(data, center):
    i = j = 0
    dim = len(data[0])
    muestra = len(data)
    suma = 0
    count = 0
    for i in range(muestra):
        suma += np.linalg.norm(data[i]-center)
        count += 1
    
    return suma/count    
```

```python
#Primero se calculan 2 clusters
clusters2 = clustering(2, arrayHistorial)
```

```python
#Se miran los promedios de distancia en cada cluster
prom_2_1 = distProm(np.array(clusters2[0][0]), clusters2[1][0])
prom_2_2 = distProm(np.array(clusters2[0][1]), clusters2[1][1])
print "Con 2 Clusters\n\nDesviación estándar del Cluster1"
print prom_2_1
print "\nDesviación estándar del Cluster2"
print prom_2_2
print "\nDesviación estándar promedio"
print (prom_2_1 + prom_2_2)/2
```

Estos numeros grandes son muy razonables ya que nos encontramos en un espacio de 24 dimensiones con 1123 puntos distribuidos mas o menos uniformemente en un espacio de rango considerable.2 clusters, probablemente tengan que tener en cuenta un conjunto de numeros distribuidos en un gran rango.

```python
#Ahora con 3 clusters
clusters3 = clustering(3, arrayHistorial)
```

```python
#Se miran los promedios de distancia en cada cluster
prom_3_1 = distProm(np.array(clusters3[0][0]), clusters3[1][0])
prom_3_2 = distProm(np.array(clusters3[0][1]), clusters3[1][1])
prom_3_3 = distProm(np.array(clusters3[0][2]), clusters3[1][2])
print "Con 3 Clusters\n\nDesviación estándar del Cluster1"
print prom_3_1
print "\nDesviación estándar de distancias del Cluster2"
print prom_3_2
print "\nDesviación estándar de distancias del Cluster3"
print prom_3_3
print "\nPromedio de desviaciones"
print (prom_3_1+prom_3_2+prom_3_3)/3
```

Aquí las desviaciones aumentaron un poco, lo que quiere decir que el conjunto de datos está mejor caracterizado como formado por 2 Clusters que por 3 (Por ahora). Las distancias siguen siendo largas, pero es normal por las dimensiones y el numero de datos.
